```python
from model import Genetic_Algorithm
from geneset import GeneSet
from tqdm import tqdm
import matplotlib.pyplot as plt
```

### Number of population Test

```python
num_generations = 100
new_gene_rate = 0.1
breed_time = 4
mutation_rate = 0.1

population = []
result = []

for pop in tqdm(range(10, 1000, 10)):
    dataset = GeneSet("att48_d.txt", pop, 1)
    model = Genetic_Algorithm(dataset)
    model_result = model.train(num_generations, new_gene_rate, breed_time, mutation_rate)
    population.append(pop)
    result.append(model_result)
plt.plot(population, result)
plt.show()
```

### Number of Generation Test

```python
num_population = 100
new_gene_rate = 0.1
breed_time = 4
mutation_rate = 0.1
dataset = GeneSet("att48_d.txt", num_population, 1)
model = Genetic_Algorithm(dataset)

generation = []
result = []
for num_gen in tqdm(range(100, 5000, 100)):
    model_result = model.train(num_gen, new_gene_rate, breed_time, mutation_rate)
    generation.append(num_gen)
    result.append(model_result)
plt.plot(generation, result)
plt.show()
```

### New generation rate Test

```python
num_population = 100
num_generation = 100
breed_time = 4
mutation_rate = 0.1
dataset = GeneSet("att48_d.txt", num_population, 1)
model = Genetic_Algorithm(dataset)

gen_rate = []
result = []
for rate in tqdm(range(10, 105, 5)):
    model_result = model.train(num_generation, rate/100, breed_time, mutation_rate)
    gen_rate.append(rate/100)
    result.append(model_result)
plt.plot(gen_rate, result)
plt.show()
```
